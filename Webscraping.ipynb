```python
!pip install requests beautifulsoup4

import requests
from bs4 import BeautifulSoup
import csv

urls = [

"https://www.busfin.uillinois.edu/",
"https://www.busfin.uillinois.edu/about",
"https://www.busfin.uillinois.edu/applications",
"https://www.busfin.uillinois.edu/forms",
"https://blogs.uofi.uillinois.edu/view/9081",
"https://www.busfin.uillinois.edu/bfpp",
"https://www.busfin.uillinois.edu/who_to_ask",

"https://www.busfin.uillinois.edu/buying_contracts/paying/vendor_invoice",
"https://www.busfin.uillinois.edu/accounting_budgeting",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/cfoapal_accounting_strings",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/cfoapal_accounting_strings/banner_fund_codes",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/cfoapal_accounting_strings/banner_account_codes",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/cfoapal_accounting_strings/banner_program_codes_NACUBO_functions",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/cfoapal_accounting_strings/terminating_banner_code_segments",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/reconciliations",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/reconciliations/month_end_closing_schedule",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/reconciliations/reconciliation_reports",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/journal_vouchers",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/payroll_accounting",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/payroll_accounting/labor_redistributions",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/payroll_accounting/fringe_benefits",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/payroll_accounting/other_payroll_errors",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/payroll_accounting/prior_year_labor_distributions",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/payroll_accounting/payroll_encumbrances",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/year_end_procedures",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/year_end_procedures/fact_sheets",
 "https://www.busfin.uillinois.edu/accounting_budgeting/accounting_financial_reporting/year_end_procedures/deadlines_for_labor_redistributions",
 "https://www.busfin.uillinois.edu/accounting_budgeting/budget",
 "https://www.busfin.uillinois.edu/accounting_budgeting/budget/budget_development",
 "https://www.busfin.uillinois.edu/accounting_budgeting/financial_controls___security",
 "https://www.busfin.uillinois.edu/accounting_budgeting/financial_controls___security/prohibited_role_combinations",
 "https://www.busfin.uillinois.edu/accounting_budgeting/financial_controls___security/prohibited_role_combinations/prohibited_role_combinations_report",
 "https://www.busfin.uillinois.edu/accounting_budgeting/financial_controls___security/prohibited_role_combinations/internal_controls_exception_requests",
 "https://www.busfin.uillinois.edu/accounting_budgeting/financial_controls___security/FCIAA",
 "https://www.busfin.uillinois.edu/accounting_budgeting/financial_controls___security/FCIAA/FCIAA_certification",
 "https://www.busfin.uillinois.edu/accounting_budgeting/financial_controls___security/fraud_prevention_awareness",
 "https://www.busfin.uillinois.edu/accounting_budgeting/financial_controls___security/fraud_prevention_awareness/reporting-fraud-misconduct-whistleblower",
 "https://www.busfin.uillinois.edu/accounting_budgeting/financial_controls___security/identity_theft_prevention_program",
 "https://www.busfin.uillinois.edu/accounting_budgeting/financial_controls___security/identity_theft_prevention_program/covered_activities",
 "https://www.busfin.uillinois.edu/accounting_budgeting/financial_controls___security/identity_theft_prevention_program/general_requirements",
 "https://www.busfin.uillinois.edu/accounting_budgeting/financial_controls___security/identity_theft_prevention_program/examples_red_flags",
 "https://www.busfin.uillinois.edu/accounting_budgeting/rate_schedules_development",
 "https://www.busfin.uillinois.edu/accounting_budgeting/rate_schedules_development/rate_schedules",
 "https://www.busfin.uillinois.edu/accounting_budgeting/rate_schedules_development/facilities_administrative_rates",
 "https://www.busfin.uillinois.edu/accounting_budgeting/rate_schedules_development/tuition_remission_rates",
 "https://www.busfin.uillinois.edu/accounting_budgeting/rate_schedules_development/fringe_benefit_rates",
 "https://www.busfin.uillinois.edu/accounting_budgeting/rate_schedules_development/federal_disclosure_statements___d_s-2_",
 "https://www.busfin.uillinois.edu/accounting_budgeting/service_and_storeroom_centers",
 "https://www.busfin.uillinois.edu/accounting_budgeting/service_and_storeroom_centers/service_center_classifications",
 "https://www.busfin.uillinois.edu/accounting_budgeting/service_and_storeroom_centers/policies_procedures_and_compliance",
 "https://www.busfin.uillinois.edu/accounting_budgeting/service_and_storeroom_centers/rate_calculation_requirements",
 "https://www.busfin.uillinois.edu/accounting_budgeting/service_and_storeroom_centers/rates",
 "https://www.busfin.uillinois.edu/accounting_budgeting/service_and_storeroom_centers/expenses",
 "https://www.busfin.uillinois.edu/accounting_budgeting/service_and_storeroom_centers/goods_or_services_sold__base_",
 "https://www.busfin.uillinois.edu/accounting_budgeting/service_and_storeroom_centers/service_activity_resources",
 "https://www.busfin.uillinois.edu/accounting_budgeting/sponsored_programs",

"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2103664",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2103678",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2101777",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2005196",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2005184",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2005248",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2005250",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2103654",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2103659",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2103692",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2116014",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2224833",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2116016",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2224848",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2005252",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2103695",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2103700",

"https://www.busfin.uillinois.edu/who_to_ask",

"https://www.busfin.uillinois.edu/buying_contracts",
"https://www.busfin.uillinois.edu/buying_contracts/procurement_laws_and_regulations",
"https://www.busfin.uillinois.edu/buying_contracts/procurement_laws_and_regulations/tax_exempt_status",
"https://www.busfin.uillinois.edu/buying_contracts/procurement_laws_and_regulations/bot_approval_thresholds_and_schedule",
"https://www.busfin.uillinois.edu/buying_contracts/procurement_laws_and_regulations/supplier_diversity",
"https://www.busfin.uillinois.edu/buying_contracts/procurement_laws_and_regulations/small_purchase_limits",
"https://www.busfin.uillinois.edu/buying_contracts/procurement_laws_and_regulations/state_agency_review_and_oversight",
"https://www.busfin.uillinois.edu/buying_contracts/procurement_laws_and_regulations/purchases_not_subject_to_the_illinois_procurement_",
"https://www.busfin.uillinois.edu/buying_contracts/procurement_laws_and_regulations/exempt_purchases",

"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/buying_specific_commodities",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/buying_specific_commodities/advertising_marketing_and_design_services",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/buying_specific_commodities/catering",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/buying_specific_commodities/computers",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/buying_specific_commodities/document_management_services",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/buying_specific_commodities/furniture",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/buying_specific_commodities/household_goods_and_equipment_relocation",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/buying_specific_commodities/office_supplies_and_equipment",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/buying_specific_commodities/office_machine_repairs",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/buying_specific_commodities/photocopiers",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/buying_specific_commodities/printing",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/buying_specific_commodities/shipping",
"https://www.busfin.uillinois.edu/buying_contracts/finding_goods_and_services/campus_stores",
 "https://www.busfin.uillinois.edu/buying_contracts/procurement_methods",
 "https://www.busfin.uillinois.edu/buying_contracts/procurement_methods/university_of_illinois_amazon_business",
 "https://www.busfin.uillinois.edu/buying_contracts/procurement_methods/renew_or_establish_annual_ibuy_orders",
 "https://www.busfin.uillinois.edu/buying_contracts/procurement_methods/creating_sealed_bid_specifications",
 "https://www.busfin.uillinois.edu/buying_contracts/procurement_methods/creating_request_for_proposal_specifications",
 "https://www.busfin.uillinois.edu/buying_contracts/procurement_methods/emergency_purchases",
 "https://www.busfin.uillinois.edu/buying_contracts/procurement_methods/sole_source_purchases",
 "https://www.busfin.uillinois.edu/buying_contracts/procurement_methods/commodity_codes",
 "https://www.busfin.uillinois.edu/buying_contracts/paying",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/pcard_tcard",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/pcard_tcard/request_card",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/pcard_tcard/roles_responsibilities",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/pcard_tcard/cardholder_guidelines",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/pcard_tcard/cardholder_guidelines/approved_vendors_for_cardholders",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/pcard_tcard/lost_stolen",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/payables_processing_status",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/vendor_setup",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/vendor_invoice",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/vendor_invoice/electronic_invoicing",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/vendor_invoice/special_payment_requests",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/vendor_invoice/receipt_required",
 "https://www.busfin.uillinois.edu/buying_contracts/paying/vendor_invoice/payment_action_requests",
 "https://www.busfin.uillinois.edu/buying_contracts/contracts",
 "https://www.busfin.uillinois.edu/buying_contracts/contracts/contract_records",
 "https://www.busfin.uillinois.edu/buying_contracts/contracts/purchase_order_terms_conditions_and_certifications",
 "https://www.busfin.uillinois.edu/buying_contracts/diverse_spend_dashboard",


"https://www.busfin.uillinois.edu/paying_people",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/payroll_schedules",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/understanding_your_paycheck",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/understanding_your_paycheck",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/understanding_your_paycheck/payroll_deductions",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/understanding_your_paycheck/payroll-distribution",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/understanding_your_paycheck/net_pay_calculations",

"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/tax_information_and_forms",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/tax_information_and_forms/form_w-2_1042-s_and_1095-c_distribution",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/tax_information_and_forms/employees_withholding_certificate",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/tax_information_and_forms/student_employee_fica_exception",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/tax_information_and_forms/waivers_departmental_payments_and_reimbursements",

"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/employee_address_change",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/payroll_transaction_processing",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/payroll_transaction_processing/payroll_payment_process_overview",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/payroll_transaction_processing/payroll_adjustment_processing",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/payroll_transaction_processing/employee_recognition_awards",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/payroll_transaction_processing/overpayments",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/payroll_transaction_processing/payroll_gross-_up_calculator",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/benefits",
"https://www.busfin.uillinois.edu/paying_people/payroll__benefits/faqs",
"https://www.busfin.uillinois.edu/paying_people/reimbursements",
"https://www.busfin.uillinois.edu/paying_people/reimbursements/miscellaneous_reimbursements",
"https://www.busfin.uillinois.edu/paying_people/reimbursements/Requesting%20Reimbursement",
"https://www.busfin.uillinois.edu/paying_people/cash_advances",
"https://www.busfin.uillinois.edu/paying_people/payments_to_foreign_nationals",
"https://www.busfin.uillinois.edu/paying_people/payments_to_foreign_nationals/foreign_national_tax_status_review",
"https://www.busfin.uillinois.edu/paying_people/payments_to_foreign_nationals/obtaining_a_temporary_control_number__tcn_",
"https://www.busfin.uillinois.edu/paying_people/payments_to_foreign_nationals/payments_to_foreign_nationals_outside_the_us",
"https://www.busfin.uillinois.edu/paying_people/payments_to_foreign_nationals/non-wage_payments",
"https://www.busfin.uillinois.edu/paying_people/payments_to_foreign_nationals/non-wage_payments/foreign_nationals_payment_eligibility_grid",

"https://www.busfin.uillinois.edu/paying_people/payments_to_non_employees",
"https://www.busfin.uillinois.edu/paying_people/payments_to_non_employees/honorarium_payments",
"https://www.busfin.uillinois.edu/paying_people/payments_to_non_employees/participant_support_costs",
"https://www.busfin.uillinois.edu/paying_people/payments_to_non_employees/Payments_to_Individuals_with_UINs",
"https://www.busfin.uillinois.edu/paying_people/payments_to_students",


"https://www.busfin.uillinois.edu/travel",
"https://www.busfin.uillinois.edu/travel/headquarters_and_travel_status",
"https://www.busfin.uillinois.edu/travel/planning_and_booking",
"https://www.busfin.uillinois.edu/travel/planning_and_booking/contracted_travel_agency",
"https://www.busfin.uillinois.edu/travel/planning_and_booking/inter-university_travel",
"https://www.busfin.uillinois.edu/travel/planning_and_booking/transportation",
"https://www.busfin.uillinois.edu/travel/planning_and_booking/transportation/passenger_vehicles_and_car_rentals",
"https://www.busfin.uillinois.edu/travel/planning_and_booking/transportation/train_and_bus_travel",
"https://www.busfin.uillinois.edu/travel/planning_and_booking/transportation/air_travel",
"https://www.busfin.uillinois.edu/travel/planning_and_booking/transportation/uber_for_business",
"https://www.busfin.uillinois.edu/travel/planning_and_booking/hotels_and_lodging",
"https://www.busfin.uillinois.edu/travel/planning_and_booking/hotels_and_lodging/hotel_discount_programs",
"https://www.busfin.uillinois.edu/travel/planning_and_booking/business-travel-meals",
"https://www.busfin.uillinois.edu/travel/international-travel",
"https://www.busfin.uillinois.edu/travel/international-travel/international_travel_insurance",
"https://www.busfin.uillinois.edu/travel/travel_payments_and_reimbursements",
"https://www.busfin.uillinois.edu/travel/business_travel_glossary",
"https://www.busfin.uillinois.edu/payments_to_the_university",

"https://www.busfin.uillinois.edu/equipment_management",
"https://www.busfin.uillinois.edu/equipment_management/equipment_overview",
"https://www.busfin.uillinois.edu/equipment_management/equipment_acquisitions",
"https://www.busfin.uillinois.edu/equipment_management/equipment_acquisitions/equipment_coding",
"https://www.busfin.uillinois.edu/equipment_management/equipment_acquisitions/tags_numbers_and_labels",
"https://www.busfin.uillinois.edu/equipment_management/equipment_acquisitions/equipment_threshold_change",
"https://www.busfin.uillinois.edu/equipment_management/maintaining_equipment_records",
"https://www.busfin.uillinois.edu/equipment_management/maintaining_equipment_records/sponsored_program_equipment",
"https://www.busfin.uillinois.edu/equipment_management/biennial_inventory",
"https://www.busfin.uillinois.edu/equipment_management/surplus_transfers_disposals",
"https://www.busfin.uillinois.edu/equipment_management/surplus_transfers_disposals/surplus_warehouse_operations",
"https://www.busfin.uillinois.edu/equipment_management/surplus_transfers_disposals/transfers_and_loans",
"https://www.busfin.uillinois.edu/equipment_management/surplus_transfers_disposals/equipment_disposal",
"https://www.busfin.uillinois.edu/equipment_management/equipment_management_glossary",

"https://www.busfin.uillinois.edu/payments_to_the_university",
"https://www.busfin.uillinois.edu/payments_to_the_university/department_receivables",
"https://www.busfin.uillinois.edu/payments_to_the_university/department_receivables/Customer_Billing",
"https://www.busfin.uillinois.edu/payments_to_the_university/department_receivables/Customer_Billing/making_payments",
"https://www.busfin.uillinois.edu/payments_to_the_university/department_receivables/department_information",
"https://www.busfin.uillinois.edu/payments_to_the_university/department_receivables/department_information/banner_AR",
"https://www.busfin.uillinois.edu/payments_to_the_university/department_receivables/department_information/billing_payments",
"https://www.busfin.uillinois.edu/payments_to_the_university/department_receivables/department_information/allowance_for_doubtful_accounts",
"https://www.busfin.uillinois.edu/payments_to_the_university/department_receivables/reconcile",

"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2139432",
"https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2139349",
 "https://www.busfin.uillinois.edu/vendors",
 "https://www.busfin.uillinois.edu/cms/One.aspx?portalId=1993898&pageId=2337977",
"https://www.busfin.uillinois.edu/payments_to_the_university/accepting_card_payments",
"https://www.busfin.uillinois.edu/payments_to_the_university/cash_handling"
]

def scrape_all_text(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        #Removes all <script>, <style>, and <noscript> tags since these contain code or irrelevant data rather than visible text
        for tag in soup(['script', 'style', 'noscript']):
            tag.extract()
        text = soup.get_text(separator=' ', strip=True) #text blocks separated by spaces, removing excess whitespace
        return text
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return ""

output_file = "uiuc_complete_content_test.csv"

with open(output_file, "w", newline='', encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["URL", "Text", "WordCount"]) #three columns

    for url in urls:
        print(f"Scraping: {url}")
        text = scrape_all_text(url)
        word_count = len(text.split())
        writer.writerow([url, text, word_count])

print(f"Scraping complete! Data saved to {output_file}")
```
